# 1번문제
# 상품 재고 관리 (딕셔너리 활용)

딕셔너리를 활용하여 상품 재고를 관리하는 프로그램을 작성하세요.
```
(결과물 , 콘솔 출력)

새로운 상품 추가
특정 상품의 재고 변경
특정 상품 삭제
전체 상품 목록 조회
특정 상품 검색
단순 기능 구현만 보이기
(예시: "사과": 10, "바나나": 5, "체리": 8)


````python
inventory = {
    "사과" : 10,
    "바나나" : 5,
    "체리" : 8
}
print(inventory)

def add(stock):
    print("새로운 상품 추가")
    input_name = input("추가할 상품 이름 : ")
    input_stock = int(input(f"{input_name}에 해당 상품의 재고 입력 : "))
    print(f"새 상품 {input_name}이 {input_stock}개 추가되었습니다.\n")
    stock[input_name] = input_stock

def num_change(stock):
    print("특정 상품의 재고 변경")
    input_name = input("재고변경할 상품 이름 : ")
    input_stock = int(input(f"{input_name}에 해당 상품 재고변경할 수량 : "))
    print(f"변경한 상품 {input_name}의 재고수량이 {input_stock}개로 변경되었습니다.\n")
    stock[input_name] = input_stock

def product_remove(stock):
    print("특정 상품 삭제")
    input_name = input("삭제할 상품의 이름 : ")
    new_stock = {name2 : number2 for name2, number2 in stock.items() if name2 != input_name}
    print(f"상품 {input_name}이 삭제되었습니다.\n")
    return new_stock

def list_search(stock):
    print("전체 상품 목록 조회")
    for name, number in stock.items():
        print(f"상품명 : {name} -> 수량: {number}\n")

def product_search(stock):
    print("특정 상품 검색")
    input_name = input("검색할 상품의 이름 : ")

    print(f"검색하신 '{input_name}' 의 재고는 {stock[input_name]}개 남았습니다.\n")

while True:
    print("재고관리 프로그램 \n")

    print("1.새로운 상품 추가")
    print("2.특정 상품의 재고 변경")
    print("3.특정 상품 삭제")
    print("4.전체 상품 목록조회")
    print("5.특정 상품 검색")
    print("6.프로그램종료")

    choice = input("원하는 번호를 선택하세요.")

    if choice == "1":
        add(inventory)
    elif choice == "2":
        num_change(inventory)
    elif choice == "3":
        product_remove(inventory)
    elif choice == "4":
        list_search(inventory)
    elif choice == "5":
        product_search(inventory)
    elif choice == "6":
        print("프로그램 종료\n")
        break

    else:print("잘못된 번호를 선택하셨습니다.")
````

# 3번문제
# 도서 관리 시스템 (상속 및 다형성 활용)
```
도서 Book , 전자책 EBook, 종이책 PaperBook 클래스를 구현하고,
display_info() 메서드를 다형성을 적용하여 다르게 동작하도록 만드세요.
--
```
참고,
(def   init  (self, title, author):,
# 객체 생성, # 정보 출력
(결과물 , 콘솔 출력)


````python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author

    def display_info(self):
        print(f"해당 도서의 제목은 {self.title}이고, 저자는 {self.author}입니다.\n")

class EBook(Book):
    def __init__(self, title, author, price):
        self.title = title
        self.author = author
        self.price = price

        super().__init__(title, author)

    def display_info(self):
        print(f"해당 도서의 제목은 {self.title}, 저자는 {self.author}, 가격은 {self.price} 입니다.\n")

class PaperBook(Book):
    def __init__(self, title, author, pages):
        self.title = title
        self.author = author
        self.pages = pages

        super().__init__(title, author)

    def display_info(self):
        print(f"해당 도서의 제목 : {self.title}, 저자 : {self.author}, 해당 책의 총 페이지수는 : {self.pages}.\n")


book_1 = Book("'빅데이터 6기'", "'이상용'")
ebook_1 = EBook("'한입크기로 잘라먹는 리액트'", "'이정환'", "'33,000'")
paperbook_1 = PaperBook("'맛있는 몽고DB'", "'정승호'", "'397p'")

book_1.display_info()
ebook_1.display_info()
paperbook_1.display_info()

books = [book_1, ebook_1, paperbook_1]
for book in books:
    book.display_info()
````

# 4번문제
# 주식 데이터(삼성전자, 카카오, 네이버) 를 활용하여 주가 변동을 시각화.

# Numpy, Matplotlib, Pandas를 활용하여 다음을 분석합니다


```
1) 주가 변동 라인 그래프
삼성전자, 카카오, 네이버의 주가 변동을 하나의 그래프로 시각화

2) 이동 평균선 추가
단기(20일) 및 장기(60일) 이동 평균선을 추가하여 주가 추세를 분석

3) 히트맵을 통한 상관관계 분석
삼성전자, 카카오, 네이버의 주가 상관관계를 히트맵으로 표현
- 주식 데이터 수집 및 처리
삼성전자(005930), 카카오(035720), 네이버(035420)의 주가 데이터를 가져오기.
- 데이터 전처리
수집한 데이터에서 종가(close) 정보를 추출하고, 이를 하나의 데이터프레임으로 결과

- 주가 변동 시각화
각 기업의 주가 변동을 라인 그래프로 시각화

- 이동 평균선 추가
주가의 단기 및 장기 추세를 파악하기 위해 20일 및 60일 이동 평균선을 계산하고 시각화

- 종목 간 상관관계 분석
각 종목 간 주가의 상관관계를 히트맵으로 시각화
```

````python
# !pip install pandas matplotlib seaborn yfinance requests_html

import time
import pandas as pd
import numpy as np
import yfinance as yf
from yahoo_fin import stock_info as si
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns

font_path = "C:/Windows/Fonts/malgun.ttf"
font_prop = fm.FontProperties(fname=font_path)
plt.rc('font', family=font_prop.get_name())
plt.rcParams['axes.unicode_minus'] = False
````

````python
# 종목 설정
tickers = {
    '삼성전자': '005930.KS',
    '네이버'  : '035420.KS',
    '카카오'  : '035720.KS'
}


HEADERS = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/124.0.0.0 Safari/537.36"
    )
}
def get_data_yfinance(ticker, start, end):
    return yf.download(ticker, start=start, end=end, progress=False, auto_adjust=True)

def get_data_safe(ticker, start="2022-01-01", end="2025-09-01"):
    try:
        df = get_data_yfinance(ticker, start, end)
        if not df.empty:
            return df
    except Exception as e:
        print(f"yfinance failed for {ticker}: {e}")

    raise RuntimeError(f"All data sources failed for {ticker}")


close_series_list = []

for company, ticker in tickers.items():
    try:
        df = get_data_safe(ticker)
        if not df.empty:

            close_series = df['Close']

            close_series.name = company

            close_series_list.append(close_series)
        else:
            print(f"[WARN] {company} ({ticker}) 데이터가 비어있습니다.")

    except Exception as e:
        print(f"[FAIL] {company} ({ticker}) -> {e}")


close_prices = pd.concat(close_series_list, axis=1)

close_prices.fillna(method='ffill', inplace=True)


csv_path = "stock_04_test.csv"
close_prices.to_csv(csv_path)
print(f"종가 데이터 저장 완료: {csv_path}")
print("생성된 DataFrame 확인:")
print(close_prices.head())


print(close_prices.head())
````

````python
# 1) 주가 변동 라인 그래프
plt.figure(figsize=(14, 7))


for company in close_prices.columns:
    plt.plot(close_prices.index, close_prices[company], label=company)

plt.title("삼성전자, 네이버, 카카오 주가 변동", fontproperties=font_prop)
plt.xlabel("날짜", fontproperties=font_prop)
plt.ylabel("종가 (원)", fontproperties=font_prop)
plt.legend(prop=font_prop)
plt.grid(True)
plt.show()
````

````python
# 2) 이동 평균선 추가
for company in close_prices.columns:

    short_ma = close_prices[company].rolling(window=20).mean()
    long_ma = close_prices[company].rolling(window=60).mean()


    plt.figure(figsize=(14, 7))
    plt.plot(close_prices.index, close_prices[company], label='종가', alpha=0.7)

    plt.plot(short_ma.index, short_ma, label='20일 이동평균선', linestyle='--')

    plt.plot(long_ma.index, long_ma, label='60일 이동평균선', linestyle=':')

    plt.title(f'{company} 주가와 이동평균선', fontproperties=font_prop)
    plt.xlabel('날짜', fontproperties=font_prop)
    plt.ylabel('종가 (원)', fontproperties=font_prop)
    plt.legend(prop=font_prop)
    plt.grid(True)
    plt.show()
````

````python
# 3) 히트맵을 통한 상관관계 분석
corr = close_prices.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm',
    vmin=-1, vmax=1,
    fmt=".2f"
)

plt.title('종목별 주가 상관관계 히트맵', fontproperties=font_prop)
plt.show()
````
